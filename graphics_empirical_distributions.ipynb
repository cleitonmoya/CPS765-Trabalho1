# Empirical Distributions - Grahphics

## Degree Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ccdf_karate = np.loadtxt('vectors/karate-degree-ccdf.txt',delimiter=',')
x_karate = np.loadtxt('vectors/karate-degree-x.txt',delimiter=',')

ccdf_metabolic_in = np.loadtxt('vectors/metabolic-degree_in-ccdf.txt',delimiter=',')
x_metabolic_in = np.loadtxt('vectors/metabolic-degree_in-x.txt',delimiter=',')
ccdf_metabolic_out = np.loadtxt('vectors/metabolic-degree_out-ccdf.txt',delimiter=',')
x_metabolic_out = np.loadtxt('vectors/metabolic-degree_out-x.txt',delimiter=',')

ccdf_powergrid = np.loadtxt('vectors/powergrid-degree-ccdf.txt',delimiter=',')
x_powergrid = np.loadtxt('vectors/powergrid-degree-x.txt',delimiter=',')

ccdf_protein= np.loadtxt('vectors/protein-degree-ccdf.txt',delimiter=',')
x_protein= np.loadtxt('vectors/protein-degree-x.txt',delimiter=',')
```

```python
plt.figure(figsize=(5,5))
plt.title('Empirical CCDF')
plt.scatter(x_karate,ccdf_karate,alpha=0.5, label='Karate')
plt.scatter(x_metabolic_in,ccdf_metabolic_in, alpha=0.5, label='Metabolic (in)')
plt.scatter(x_metabolic_out,ccdf_metabolic_out, alpha=0.5, label='Metabolic (out)')
plt.scatter(x_powergrid,ccdf_powergrid,alpha=0.5, label='Powergrid')
plt.scatter(x_protein,ccdf_protein,alpha=0.5, label='Protein')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Degree')
plt.ylabel('Complimentary Cumulative Probability')
plt.ylim(10**-4, 10**0)
plt.legend()
plt.savefig('degree.pdf')
```

## Distance Distribution

```python
x_karate_dist = np.loadtxt('vectors/karate-dist-x.txt',delimiter=',')
pmf_karate_dist = np.loadtxt('vectors/karate-dist-pmf.txt',delimiter=',')
ccdf_karate_dist = np.loadtxt('vectors/karate-dist-ccdf.txt',delimiter=',')

x_metabolic_dist = np.loadtxt('vectors/metabolic-dist-x.txt',delimiter=',')
pmf_metabolic_dist = np.loadtxt('vectors/metabolic-dist-pmf.txt',delimiter=',')
ccdf_metabolic_dist = np.loadtxt('vectors/metabolic-dist-ccdf.txt',delimiter=',')

x_powergrid_dist = np.loadtxt('vectors/powergrid-dist-x.txt',delimiter=',')
pmf_powergrid_dist = np.loadtxt('vectors/powergrid-dist-pmf.txt',delimiter=',')
ccdf_powergrid_dist = np.loadtxt('vectors/powergrid-dist-ccdf.txt',delimiter=',')

x_protein_dist = np.loadtxt('vectors/protein-dist-x.txt',delimiter=',')
ccdf_protein_dist = np.loadtxt('vectors/protein-dist-ccdf.txt',delimiter=',')
pmf_protein_dist = np.loadtxt('vectors/protein-dist-pmf.txt',delimiter=',')
```

### Empirical Distribution:

```python
plt.figure(figsize=(5,5))
plt.title('Empirical Distribution')

plt.scatter(x_karate_dist, pmf_karate_dist, alpha=0.5)
plt.plot(x_karate_dist, pmf_karate_dist, alpha=0.5, label='Karate')

plt.scatter(x_metabolic_dist, pmf_metabolic_dist, alpha=0.5)
plt.plot(x_metabolic_dist, pmf_metabolic_dist, alpha=0.5, label='Metabolic')

plt.scatter(x_powergrid_dist, pmf_powergrid_dist, alpha=0.5)
plt.plot(x_powergrid_dist, pmf_powergrid_dist, alpha=0.5, label='Powergrid')

plt.scatter(x_protein_dist, pmf_protein_dist, alpha=0.5)
plt.plot(x_protein_dist, pmf_protein_dist, alpha=0.5, label='Protein')

plt.xlabel('Distace')
plt.ylabel('Probability')
plt.xlim(0,30)
plt.legend()
plt.savefig('distance.pdf')
```
